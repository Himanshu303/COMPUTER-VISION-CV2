# MANIPULAITNG IMAGES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
img=cv2.imread("Images/test2.jpg",0)
plt.imshow(img,cmap='gray')
print(img)
```

```python
#ACESSSING PIXELS
print(img[0][0])
print(img[0][300])
```

```python
#MODIFYING PIXELS
imcop=img.copy()
imcop[0][300]=0
imcop[0][299]=0
imcop[0,301:310]=135
imcop[100:201,0:100]=135

plt.imshow(imcop,cmap='gray')
```

# CROPPING IMAGES

```python
turt=cv2.imread("Images/test3.jpg",cv2.IMREAD_COLOR)
#cv2.imread reads image in BGR form 
turt_rgb=turt[:,:,::-1]
#by this we convert BGR TO RGb form
plt.imshow(turt_rgb)
```

```python
cropped=turt_rgb[200:400,650:850]
plt.imshow(cropped)
```

```python
#RESIZING IMAGE with dsize as None
cropped_2x=cv2.resize(cropped,None,fx=2,fy=2)
plt.imshow(cropped_2x)
```

```python
#RESIZING WITHOUT MAINTAINING ASPECT RATIO
width=100
height=200
dim=(width,height)
newcrop=cv2.resize(cropped,dsize=dim,interpolation=cv2.INTER_AREA)
plt.imshow(newcrop)
```

```python
#RESIZING WITH MAINTAINING ASPECT RATIO
asp_rat=width/cropped.shape[1]
height=int(cropped.shape[0]*asp_rat)
dim=(width,height)

newcrop1=cv2.resize(cropped,dsize=dim)
plt.imshow(newcrop1)
```

```python
#FLIPPING IMAGE HORIZONTALLY

plt.imshow(cv2.flip(turt,1))
```

```python
#FLIPPING IMAGE VERTICALLY
plt.imshow(cv2.flip(turt,0))
```

```python
#FLIPPING IMAGE HORIZONTALLY AS WELL AS VERTICALLY
plt.imshow(cv2.flip(turt,-1))
```
