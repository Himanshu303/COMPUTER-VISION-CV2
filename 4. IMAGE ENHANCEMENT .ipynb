# IMAGE ENHANCEMENET USING MATHEMATICAL OPERATIONS

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
from PIL import Image
import matplotlib
matplotlib.rcParams['figure.figsize']=(9.0,9.0)
```

# ORIGINAL IMAGE

```python
img=cv2.imread("Images/test5.jpg",cv2.IMREAD_COLOR)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

# Adjusting Brightness

```python
matrix=np.ones(img.shape,dtype="uint8")*50

img_darker=cv2.subtract(img,matrix)
img_brighter=cv2.add(img,matrix)

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_darker); plt.title("DARKER")
plt.subplot(132); plt.imshow(img); plt.title("ORIGINAL")
plt.subplot(133); plt.imshow(img_brighter); plt.title("BRIGHTER")
```

# Adjusting Contrast

```python
matrix1=np.ones(img.shape)*0.7
matrix2=np.ones(img.shape)*1.3

img_darker=np.uint8(cv2.multiply(np.float64(img),matrix1))
img_brighter=np.uint8(cv2.multiply(np.float64(img),matrix2))

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_darker); plt.title("LOWER CONTRAST")
plt.subplot(132); plt.imshow(img); plt.title("ORIGINAL")
plt.subplot(133); plt.imshow(img_brighter); plt.title("HIGHER CONTRAST")
```

# Clipping values to avoid overflow

```python
matrix1=np.ones(img.shape)*0.70
matrix2=np.ones(img.shape)*1.3

img_darker=np.uint8(cv2.multiply(np.float64(img),matrix1))
img_brighter=np.uint8(np.clip(cv2.multiply(np.float64(img),matrix2),0,255))

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_darker); plt.title("LOWER CONTRAST")
plt.subplot(132); plt.imshow(img); plt.title("ORIGINAL")
plt.subplot(133); plt.imshow(img_brighter); plt.title("HIGHER CONTRAST")
```

# Image Thresholding

```python
img=cv2.imread("IMAGES/test6.jpg",cv2.IMREAD_GRAYSCALE)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)
```

```python
retval,img_th=cv2.threshold(img,140,255,cv2.THRESH_BINARY)
plt.figure(figsize=[18,5])
plt.subplot(121); plt.imshow(img);plt.title("ORIGINAL")
plt.subplot(122); plt.imshow(img_th);plt.title("THRESHOLDING:140")
```

```python
img=cv2.imread("IMAGES/test6.jpg",cv2.IMREAD_GRAYSCALE)
retval,img_th1=cv2.threshold(img,75,255,cv2.THRESH_BINARY)

retval,img_th2=cv2.threshold(img,130,255,cv2.THRESH_BINARY)

img_adp=cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,7)

plt.figure(figsize=[18,15])
plt.subplot(221); plt.imshow(img,cmap='gray');plt.title("ORIGINAL")
plt.subplot(222); plt.imshow(img_th1,cmap='gray');plt.title("THRESHOLDING:75")
plt.subplot(223); plt.imshow(img_th2,cmap='gray');plt.title("THRESHOLDING:130")
plt.subplot(224); plt.imshow(img_adp,cmap='gray');plt.title("THRESHOLDING:ADAPTIVE")
```
