# ANNOTATING IMAGES

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
from PIL import Image
import matplotlib
matplotlib.rcParams['figure.figsize']=(9.0,9.0)
```

```python
img=cv2.imread("Images/test4.jpg",1)
plt.imshow(img[:,:,::-1])
```

```python
#DRAWING A LINE

imcopy=img.copy()
cv2.line(imcopy,(300,275),(375,275),(255,255,255),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(imcopy[:,:,::-1])
```

```python
#DRAWING A CIRCLE

imcopy=img.copy()
cv2.circle(imcopy,(332,215),75,(0,0,255),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(imcopy[:,:,::-1])
```

```python
#DRAWING A RECTANGLE

imcopy=img.copy()
cv2.rectangle(imcopy,(400,300),(550,150),(0,0,255),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(imcopy[:,:,::-1])
```

```python
imcopy=img.copy()
text="NEPTUNE"
fontScale=2.3
fontFace=cv2.FONT_HERSHEY_PLAIN
fontColor=(0,255,0)
fontThickness=2
cv2.rectangle(imcopy,(800,250),(920,150),(0,0,255),thickness=5,lineType=cv2.LINE_AA)
cv2.putText(imcopy,text,(780,300),fontFace,fontScale,fontColor,fontThickness,cv2.LINE_AA)
plt.imshow(imcopy[:,:,::-1])
```
